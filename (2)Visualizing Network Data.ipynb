## Visualizing Network Data

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Create and draw simple path graph
G1 = nx.path_graph(8)
nx.draw(G1, with_labels = True)
plt.show()
```

```python
# Create and draw star graph
G2 = nx.star_graph(10)
nx.draw(G2, with_labels = True)
plt.show()
```

```python
# Create and draw complete graph
G3 = nx.complete_graph(20)
nx.draw(G3, with_labels = True)
plt.show()
```

## Adjacency Matrix

```python
# Create an empty graph with no nodes and no edges using the Graph() function
G4 = nx.Graph()

# Add multiple nodes to the graph using add_node_from
G4.add_nodes_from(['A', 'B', 'C', 'D'])

# Add multiple edges at once
G4.add_edges_from([('A', 'B'), ('A', 'C'), ('A', 'D')])
```

```python
# Plot graph
nx.draw(G4, with_labels = True)
plt.show()
```

```python
# adjacency matrix (returns a SciPy sparse matrix)
A = nx.adjacency_matrix(G4)
print(A.todense())
```

```python
# adjacency matrix (as a Numpy Matrix)
A_np = nx.to_numpy_matrix(G4)
A_np
```

```python
# adjacency matrix (as a Pandas Dataframe)
A_pd = nx.to_pandas_adjacency(G4)
A_pd
```

## Incidence Matrix

```python
# incidence matrix (returns a SciPy sparse matrix)
I = nx.incidence_matrix(G4)
print(I.todense())
```
